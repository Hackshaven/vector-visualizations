# Test Client for Tensor Server

Accepts from following API calls...

* URL/ -- Simple welcome message. Add Help text in future
* URL/tensor -- List the tensor files available, along with info (like shape)
* URL/tensor?name=asdf -- Dump the value of the 'asdf' tensor, as binary

```python
import requests
```

## Return Help  on Using the Tensor Server

```python
url = "http://localhost:5000/"
r = requests.get(url) 
```

```python
if r.status_code == 200:
    print(f"Success: {r.text}")
else:
    print(f"Failure: {r.text}")
```

## Return List of Available Tensors as JSON

```python
r = requests.get(url + '/tensor')
```

```python
if r.status_code == 200:
    print(f"Success: {r.text}")
else:
    print(f"Failure: {r.text}")
```

## Return Dump of Values for Specific Tensor

```python
name = 'uvw_test16.uvw'
r = requests.get(url + '/tensor' + '?name=' + name)
```

```python
if r.status_code == 200:
    print(f"Success: {r.text}")
else:
    print(f"Failure: {r.text}")
```
